#### Part 1
-- 1) create a function that will generate 'n' weather data records. The schema is as follows:

> **NO HARDCODING**


- rowid: unique record identifier
- cityname: unique city name
- mintemp: average minimum temperature (random integers between 5-15) 
- maxtemp: average maximum temperature (random integers between 25-35) 
- minhumid: average minimum humidity (random integers between 1-100)
- maxhumid: average maximum humidity (random integers between 1-100) 
- alt: Altitude (random integers between 200-4000)
- pollution: Pollution level (random float numbers between 1-50; 1-decimal precision)
- wind dir: Wind direction (any 1 value from 'n', 's', 'w', 'e', 'ne', 'sw') 
- windspeed: Wind speed (random integers between 25-150)
- atmpress: Atmospheric pressure (random float between 400-1200 with 1-decimal precision)

- arealoc: Location (any 1 value from 'high', 'medium', 'low') 
- floods: Flood conditions (any 1 value from 'always', 'rarely', 'never')
- visibility: Average visibility (random float between 0.25-10)
- acc_due_to_fog: Accidents due to fog (random integers between 1000-15000)

```python
# importing basic libraries
import numpy as np
import pandas as pd
import random
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
class Weather1:
  def __init__(self):
    self.cities = ["Mumbai","Delhi","Bangalore","Hyderabad",
                  "Ahmedabad","Chennai","Kolkata","Surat","Vadodara","Pune"]

  def weather_data(self):
    self.mintemp = np.random.randint(5,16,len(self.cities))                 # average minimum temperature (random integers between 5-15
    self.max_temp = np.random.randint(25,36,len(self.cities))                     # average maximum temperature (random integers between 25-35)
    self.minhumid = np.random.randint(1,101,len(self.cities))               # average minimum humidity (random integers between 1-100)
    maxhumid = np.random.randint(1,101,len(self.cities))
    self.alt = np.random.randint(200,4001,len(self.cities))                 # Altitude (random integers between 200-4000)
    self.windspeed = np.random.randint(25,151,len(self.cities))             # Wind speed (random integers between 25-150)
    self.acc_due_to_fog = np.random.randint(1000,15001,len(self.cities))    # Accidents due to fog (random integers between 1000-15000)

    # maximum humidity which should be gretter than mintemp
    max_humid=[]
    for k,l in zip(self.minhumid,maxhumid):
      if l>k :
        max_humid.append(l)
      else:
        max_humid.append(k)


    self.maxhumid = max_humid                                             # average maximum humidity (random integers between 1-100)                                       

    self.pollution = [];self.wind_dir=[];self.arealoc = [];self.visibility=[]
    self.atmpress =[];self.floods = []
    area_l = ['high', 'medium', 'low']
    wd = ['n', 's', 'w', 'e', 'ne', 'sw']
    fd = ['always', 'rarely', 'never']
    self.index = []
    for i in range(1,len(self.cities)+1):
      self.index.append(i)
      self.pollution.append(round(random.uniform(1, 52), 1))          # random float numbers between 1-50; 1-decimal precision
      self.wind_dir.append(random.choice(wd))                         # Wind direction (any 1 value from 'n', 's', 'w', 'e', 'ne', 'sw')
      self.atmpress.append(round(random.uniform(400, 1200), 1))       # Atmospheric pressure (random float between 400-1200 with 1-decimal precision)
      self.arealoc.append(random.choice(area_l))                      # Location (any 1 value from 'high', 'medium', 'low')
      self.floods.append(random.choice(fd))                           # Flood conditions (any 1 value from 'always', 'rarely', 'never')
      self.visibility.append(round(random.uniform(0.25, 10), 2))      # Average visibility (random float between 0.25-10)


    self.dataframe = pd.DataFrame({"Row_id":self.index,"Cityname":self.cities,
                    "Mintemp":self.mintemp,
                    "Maxtemp":self.max_temp,
                    "Minhumid":self.minhumid,
                    "Maxhumid":self.maxhumid,
                    "Altitude":self.alt,
                    "Pollution":self.pollution,
                    "Wind_dir":self.wind_dir,
                    "Windspeed":self.windspeed,
                    "Atmpress":self.atmpress,
                    "Arealoc":self.arealoc,
                    "Floods":self.floods,
                    "Visibility":self.visibility,
                    "Acc_due_to_fog":self.acc_due_to_fog})

    return self.dataframe
```

##### 11) Do a statistical analysis on the numeric data and identify outliers

```python
 # defining function for outlier treatment

def out_liers (new_df, j):
    q3=new_df[j].quantile (0.75) 
    q1=new_df[j].quantile(0.25)
    iqr = q3-q1 
    
    lower_fence = q1-1.5 * iqr
    upper_fance = q3+1.5 * iqr 
    m=new_df[j].median ()
    for index, x in enumerate(new_df[j]):
        if (x < lower_fence) or (x > upper_fance): 
            new_df.loc[index, j] = m
```
